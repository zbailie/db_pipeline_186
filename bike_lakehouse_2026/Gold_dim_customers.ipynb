# Business Transformation and Modeling

```python
query = """
SELECT
    ROW_NUMBER() OVER (ORDER BY ci.customer_id) AS customer_key,
    ci.customer_id,
    ci.first_name,
    ci.last_name,
    ci.marital_status,
    ci.created_date AS create_date
FROM silver.crm_customers ci
"""
df = spark.sql(query)
```

# Write to Gold Table

```python
(
    df.write
     .mode("overwrite")
     .format("delta")
     .saveAsTable("gold.dim_customer")
)
```
